```python
#Importing Packages
from secrets import choice
import speech_recognition
import pyttsx3
from tkinter import *
import tkinter.font as font
import pandas as pd
import random

#Speech Modules
engine = pyttsx3.init('sapi5')
voices = engine.getProperty('voices')
engine.setProperty('voice',voices[1].id)

def speak(text):
    engine.setProperty("rate",150)
    engine.say(text)
    engine.runAndWait()


#Window
root = Tk()
root.geometry("1280x853")
root.title("Guruvedic 2020")
root.grid()

filename = PhotoImage(file = "phot.png")
background_label = Label(root, image=filename)
background_label.place(x=0, y=0, relwidth=1, relheight=1)


# Fonts
myfont = font.Font(size=100)
myfont1 = font.Font(family="Helvetica",size=100,slant="italic",weight="bold")
myfont2 = font.Font(size=20)
myfont3 = font.Font(size=120)
myfont4 = font.Font(family="Times",size=30,slant="italic")

#Methods
def randnum(event):
  data = pd.read_csv('CSE Faculty.csv',encoding= 'unicode_escape')
  df = pd.DataFrame(data)
  arr = df['EMP ID WITH NAMES'].tolist()
  value = choice(arr)
  print(value)
  #speak(value)
  updateDisplay(value)

def updateDisplay(myString):
  displayVariable.set(myString)


#Display
b=root.config()
space1 = Label(root,font=myfont2)
space2 = Label(root,font=myfont3)
button_1 = Button(root, text="Click Here to give Faculty ID:) ",bg='white',font=250)
displayVariable = StringVar()

displayLabel = Label(root, textvariable = displayVariable,font=myfont4)
button_1.bind("<Button-1>",randnum)
space1.pack()
space2.pack()
space2.pack()
space2.pack()
displayLabel.pack()
button_1.pack()
root.mainloop()
```
